Doubly Linked list

```python
class Node:
    def __init__(self,data):
        self.data = data
        self.next = None
        self.back = None
def arr_to_dll(a):
    head = Node(a[0])
    prev = head
    for i in a[1:]:
        ne = Node(i)
        ne.back = prev
        prev.next = ne
        prev = ne
    return head
def f_traves(head):
    if not head:
        return
    move  = head
    while move:
        print(move.data,end="-->")
        move = move.next
    print("None")

def b_traves(head):
    if not head:
        return
    move = head
    while move.next:
        move = move.next
    while move:
        print(move.data,end="-->")
        move = move.back
    print("None")


a = [6,1,4,8]
head= arr_to_dll(a)
f_traves(head)
b_traves(head)
```
