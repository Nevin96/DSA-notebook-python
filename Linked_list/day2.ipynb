```python
class Node:
    def __init__(self,data):
        self.data = data
        self.next = None
        self.back = None
def arr_to_dll(a):
    head = Node(a[0])
    prev = head
    for i in a[1:]:
        ne = Node(i)
        ne.back = prev
        prev.next = ne
        prev = ne
    return head
def tra(head):
    if not head:
        return
    move = head
    while move:
        print(move.data,end="-->")
        move = move.next
    print("None")
def dlt_tail(head):
    if not head:
        return None
    if not head.next:
        return None
    move = head
    while move.next:
        move = move.next
    move.back.next = None
    return head
def dlt_kth(head,ind):
    if not head:
        return
    if ind ==1 :
        move = head
        head = head.next
        head.back = None
        move.next = None
        return head
    move = head
    count = 1
    while move and count < ind:
        count +=1
        move = move.next
    prev = move.back
    front = move.next
    if prev:
        prev.next = front
    if front:
        front.back = prev

    move.back = move.next = None
    return head
        
ar = [6,0,4,8]
head = arr_to_dll(ar)
tra(head)
# c = dlt_tail(head)
d = dlt_kth(head,4)
tra(d)
```

```python
def dlt_kth_element(head,ind):
    if not head:
        return None
    if head.next == None:
        return head
    if ind == 1:
        move = head
        head = head.next
        head.back = None
        move.next = None
        return head
    move = head
    count = 1
    while move and count < ind:
        count+=1
        move = move.next
    prev = move.back
    front = move.next
    if prev:
        prev.next = front
    if front:
        front.back = prev
    move.next = move.back = None
    return head
    
wq = [1,6,4,3]
hea = arr_to_dll(wq)
tra(hea)
c = dlt_kth_element(hea,4)
tra(c)
```

```python
def dlt_k_val(head,k):
    if not head:
        return None
    
    if k == head.data:
        if head.next == None:
            return None
        move = head
        head = head.next
        move.next = None
        head.back = None
        return head
    move = head
    while move:
        if move.data == k:
            break
        move = move.next
    prev = move.back
    front = move.next
    if prev:
        prev.next = front
    if front:
        front.back = prev
    move.next = move.back = None
    return head

wq = [1,6,4,3]
hea = arr_to_dll(wq)
tra(hea)
c = dlt_k_val(hea,1)
tra(c)
```
